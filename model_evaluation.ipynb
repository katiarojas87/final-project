# Model evalutions
Compare mean squared errors between different models

## Import packages and data

```python
import pandas as pd
import pathlib
from sklearn.metrics import root_mean_squared_error
```

```python
data_path = pathlib.Path("./data_dump")
data = pd.read_csv(data_path / "listings_with_scores.csv")
```

## Base models

### Base model: Mean price

```python
data["pred"] = data["price_man_yen"].mean()

mse_mean_price = root_mean_squared_error(data["price_man_yen"], data["pred"])
mse_mean_price
```

### Base model: Mean price per nearest station

```python
mean_score = data.groupby(['nearest_station'])["price_man_yen"].mean().reset_index()
mean_score = mean_score.rename(columns={'price_man_yen': 'avg_price_per_station'})
data = data.merge(mean_score, on="nearest_station")

mse_mean_per_station = root_mean_squared_error(data["price_man_yen"], data["avg_price_per_station"])
mse_mean_per_station
```
