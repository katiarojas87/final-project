```python
from sklearn.metrics import log_loss, accuracy_score
import pandas as pd
import json
```

```python
df = pd.read_csv('assessment_output.csv', sep=";")
df.head()
```

```python
df = df.drop(columns=['listing_url', 'image_url', "image_name", 'default_image'])
df.head()
```

```python
json_string = df["scoring_dict"]
scoring_dict = pd.json_normalize(json_string.apply(eval))
df["score_lux"] = scoring_dict["luxury"]
df["score_bright"]= scoring_dict["brightness"]
df["score_cond"]= scoring_dict["condition"]
df.head()
```

```python
keep_rows = df["room_type"] != "floor plan"
df_cleaned = df[keep_rows]
df_cleaned.head()
```

```python
#log_loss
loss_lux = log_loss(df_cleaned["Luxury"] == "yes", df_cleaned["score_lux"])
loss_bright = log_loss(df_cleaned["Brightness"] == "yes", df_cleaned["score_bright"])
loss_cond = log_loss(df_cleaned["Condition"] == "new", df_cleaned["score_cond"])
loss_lux, loss_bright, loss_cond
```

```python
#accuracy
lux_pred = df_cleaned["score_cond"] > 0.5
lux_actual = df_cleaned["Condition"] == "yes"
accuracy = accuracy_score(lux_pred, lux_actual)
accuracy
```
